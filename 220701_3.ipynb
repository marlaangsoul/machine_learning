### 첫 번째 머신러닝 프로그램
- 머신러닝 패키지 : 사이킷런 scikit-learn
- k-최근접이웃 k-Nearest Neighbors 알고리즘 사용
- 사이킷런 훈련 데이터의 shape 2차원의 특성 리스트 필요
- 도미 35마리와 빙어 14마리 구분하기


```python
a =  [1,2,3]
b = [10,20,30]
print(list(zip(a,b)))
print(dict(list(zip(a,b))))
```

```python
lst = list(zip(a,b))
lst
```

```python
# 도미와 빙어 합치기
length = bream_length+smelt_length
weight = bream_weight+smelt_weight
```

```python
fish_data = [[l,w] for l,w in list (zip(length, weight))]
fish_data
```

```python
# 정답 준비 : 도미 1, 빙어 0
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
# k-최근접이웃
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 객체(모델) 생성
kn = KNeighborsClassifier()
```

```python
# 훈련 training : 모델에 데이터를 전달하여 규칙을 학습하는 과정
kn.fit(fish_data, fish_target)
```

```python
# 평가 : 0 ~ 1 
kn.score(fish_data, fish_target)
```

### 최근접 이웃 알고리즘
- 주위의 데이터를 보고 다수를 차지하는 것을 값으로 판단
- 새로운 데이터 예측시 가장 가까운 직선거리로 판단,주위란 각 데이터의 벡터 거리
- 모든 훈련 데이터를 가지고 있어야 함
- k는 판단을 위한 주위의 가장 가까운 데이터 갯수

#### (length,height) : (30, 600)은 어떤 생선인가?

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(16, 100, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# predict() 훈련된 모델을 가지고 실제  데이터 판단 또는 예측 
#  1: 도미, 0: 빙어
kn.predict([[30, 600]])
```

위의 데이터는 도미다. 

```python
# fish_data 모두 갖고 있음
print(kn._fit_X)
```

```python
# n_neighbors : k 갯수 , 기본값 5
#  49개를 이면 모두 도미로 예측, 도미가 35 > 빙어 14
kn49 = KNeighborsClassifier(n_neighbors=49)
# 데이터가 들어 왔을때 가장 가까운 데이터 49개를 찾고 
# 그 중 더많은 근접데이터(여기서는 도미)를 정답으로 인식한다. 
# 새로 훈련
kn49.fit(fish_data, fish_target)
```

```python
# predict() 훈련된 모델을 가지고 실제  데이터 판단 또는 예측 
#  1: 도미, 0: 빙어
kn.predict([[9.8, 6.7]]),kn49.predict([[9.8, 6.7]])
```

```python
# 평가
kn49.score(fish_data, fish_target)
```

## 도미와 빙어 구분하기(분류) 문제해결 과정
1. 도미 35마리와 빙어 14마리 길이와 무게를 측정해서 파이선 리스트 생성
2. 도미와 빙어 데이터를 합쳐 2차원 리스트 준비
3. 사이킷런의 분류 알고리즘 중 k-최근접이웃 알고리즘으로 훈련
4. 새로운 생성 데이터 예측
5. 결과: 100% 예측 성공

#### 확인 문제
- n_neighbors 설정값을 변경해서 정확도가 1이 아닌 k값을 찾아보기

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```
